# Logistic Regression 
Which variables have the strongest impact on a successful customer transaction (Revenue column)?

Workflow:
Separated X & Y
TTS
Z-scored
Logistic Regression Model
Predicted Vals
Accuracy Score
Confusion Matrix 
Coefs in Odds

```python
predictors = ['Administrative','Administrative_Duration', 'Informational', 'Informational_Duration',
              'ProductRelated',
              'ProductRelated_Duration', 'BounceRates', 'ExitRates', 'PageValues']

x = df[predictors]
Y = df["Revenue"]


# tts
X_train, X_test, y_train, y_test = train_test_split(x, Y, test_size=0.2)
```

```python
# z-score
zscore = StandardScaler()
zscore.fit(X_train)
Xz_train = zscore.transform(X_train)
Xz_test = zscore.transform(X_test)
```

```python
# regression model 

logit = LogisticRegression()
```

```python
# fit 

logit.fit(Xz_train,y_train)
```

```python
# predicted values 

predictedVals = logit.predict(Xz_test) #predict
```

```python
# acc score 

accuracy_score(y_test,predictedVals)
```

```python
plot_confusion_matrix(logit,X_test, y_test)
plt.title('Confusion matrix of the classifier')
```

```python
# coef df
coef = pd.DataFrame({"Coefs": logit.coef_[0],
                     "Names": predictors})
coef = coef.append({"Coefs": logit.intercept_[0],
                    "Names": "intercept"}, ignore_index = True)
coef
```

```python
ggplot(coef, aes(x='Names', y='Coefs', fill='Names')) + geom_bar(stat='identity') + theme_minimal() + labs(title='Coefficients') + theme(axis_text_x  = element_text(angle = 90, hjust = 1))
```

```python
coef["Odds Coefs"] = np.exp(coef["Coefs"])
coef
```

```python
ggplot(coef, aes(x='Names', y='Odds Coefs')) + geom_bar(stat='identity', fill='blue') + theme_minimal() + labs(title='Coefficients in terms of Odds') + theme(axis_text_x  = element_text(angle = 90, hjust = 1))
```

# Clustering 
Are there any definitive clusters between the ExitRate and BounceRate columns? If so, are there any outliers? 

Workflow:
Create DF 
Z-scored 
Mins & Eps using Elbow Method
DBSCAN
Silhouette Score 

```python
# new df 

new = df[['BounceRates','ExitRates']]
new.head()
```

```python
# z-score

znew = StandardScaler()
new[["BounceRates","ExitRates"]] = znew.fit_transform(new)

ggplot(new, aes(x = "BounceRates", y = "ExitRates")) + geom_point() + theme_minimal()
```

```python
# we ask for mins + 1 nearest, because the data point it self (distance = 0) is included
mins = 3
nn = NearestNeighbors(mins + 1)

nn.fit(new[["BounceRates", "ExitRates"]])

distances, neighbors = nn.kneighbors(new[["BounceRates", "ExitRates"]])

distances
```

```python
# sort the distances
distances = np.sort(distances[:, mins], axis = 0)
```

```python
#plot the distances
distances_df = pd.DataFrame({"distances": distances,
                             "index": list(range(0,len(distances)))})
plt = (ggplot(distances_df, aes(x = "index", y = "distances")) +
 geom_line(color = "white", size = 2) + theme_minimal() +
 labs(title = "Elbow Method for Choosing eps") +
 theme(panel_grid_minor = element_blank(),
      rect = element_rect(fill = "#202124ff"),
      axis_text = element_text(color = "white"),
      axis_title = element_text(color = "white"),
      plot_title = element_text(color = "white"),
      panel_border = element_line(color = "darkgray"),
      plot_background = element_rect(fill = "#202124ff")
      ))
ggsave(plot=plt, filename='elbow.png', dpi=300)

plt
```

```python
# dbscan and plot

db1 = DBSCAN(eps = 0.50, min_samples = 150).fit(new)

labsList = ["Noise"]
labsList = labsList  + ["Cluster " + str(i) for i in range(1,len(set(db1.labels_)))]

new["assignments"] = db1.labels_

(ggplot(new, aes(x = "BounceRates", y = "ExitRates", color = "factor(assignments)")) +
 geom_point() +
 theme_minimal() +
 scale_color_discrete(name = "Cluster Assignment",
                      labels = labsList) +
theme(panel_grid_major = element_blank()) +
labs(title = "DBSCAN with eps = 0.5, min_samples = 150"))
```

```python
# grab only clustered data points
clustered = new.loc[(new.assignments >= 0)]

print("clustered silhouette score:", silhouette_score(clustered[["BounceRates","ExitRates"]], clustered["assignments"]))

# overall 
print("overall silhouette score:", silhouette_score(new[["BounceRates","ExitRates"]], new["assignments"]))
```

# PCA
Are there any methods to reduce the dimensionality of the continuous variables in your data set? If so, which method, how can you tell, and how many variables do you need to 
retain 80% of the original variance?

Workflow:
Selected predictor columns 
Z-scored
PCA
DF of PC's
Scree & 80% Variance Plot
Logistic Regression
PC score

```python
print(df.columns)
df.head()
```

```python
features = ['Administrative', 'Administrative_Duration', 'Informational', 
             'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
             'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay'] # grabbed all numeric columns of interest
df.head()

z = StandardScaler()
df[features] = z.fit_transform(df[features])
```

```python
# create PCA obj 
pca = PCA()
pca.fit(df[features])
```

```python
pcaDF = pd.DataFrame({"expl_var" :
                      pca.explained_variance_ratio_,
                      "pc": range(1,11), # array length 
                      "cum_var":
                      pca.explained_variance_ratio_.cumsum()})
pcaDF.head()
```

```python
# plot
(ggplot(pcaDF, aes(x = "pc", y = "expl_var")) + geom_line() + geom_point()) + theme_bw() + labs(title = "Scree Plot")
```

```python
# 80% variance 
(ggplot(pcaDF, aes(x = "pc", y = "cum_var")) + geom_line(color = "pink") +
 geom_point(color = "pink") + geom_hline(yintercept = 0.80)) + theme_bw() + labs(title = "80% of Original Variance")
```

```python
pcomps4 = pca.transform(df[features])
pcomps4 = pd.DataFrame(pcomps4[:,0:4])


pcomps6 = pca.transform(df[features])
pcomps6 = pd.DataFrame(pcomps6[:, 0:6])

#modeMod1
lr1 = LogisticRegression()
lr1.fit(df[features], df["Revenue"])
print("all data: ", lr1.score(df[features], df["Revenue"]))

#modeMod1
lr2 = LogisticRegression()
lr2.fit(pcomps6, df["Revenue"])
print("6 PCs:    ", lr2.score(pcomps6, df["Revenue"]))

#modeMod1
lr3 = LogisticRegression()
lr3.fit(pcomps4, df["Revenue"])
print("4 PCs:    ", lr3.score(pcomps4, df["Revenue"]))
```
